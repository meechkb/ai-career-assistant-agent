# AI Career Assistant Agent

This project is a simple AI agent-style workflow that analyzes a job description and generates career guidance.

The agent produces:
1. A role summary  
2. Key required skills  
3. Resume keyword suggestions  
4. Interview prep questions  
5. An application strategy  

This project was built to practice AI agent prompting, structured output generation, workflow design, and documentation.

```python
# AI Career Assistant Agent
# This version does not require an API key.
# It uses keyword matching and structured logic to simulate an AI agent workflow.

def analyze_job_description(job_text):
    job_text_lower = job_text.lower()

    skills = []

    keyword_map = {
        "python": "Python programming",
        "javascript": "JavaScript",
        "api": "APIs and integrations",
        "webhook": "Webhooks",
        "ai": "Artificial Intelligence",
        "machine learning": "Machine Learning",
        "chatbot": "Chatbots / Conversational AI",
        "prompt": "Prompt engineering",
        "workflow": "Workflow design",
        "git": "Git / version control",
        "documentation": "Technical documentation",
        "client": "Client communication",
        "testing": "Testing and debugging",
        "spreadsheet": "Spreadsheets and knowledge organization",
        "knowledge base": "Knowledge base organization"
    }

    for keyword, skill in keyword_map.items():
        if keyword in job_text_lower:
            skills.append(skill)

    if len(skills) == 0:
        skills.append("General problem-solving and communication")

    role_summary = (
        "This role appears to focus on building, testing, and improving AI-powered workflows "
        "that solve real problems for users or clients."
    )

    resume_keywords = skills[:]

    interview_questions = [
        "Can you describe a technical project you built from start to finish?",
        "How would you test an AI agent before giving it to a real user?",
        "How do you explain technical concepts to non-technical people?",
        "What steps would you take if an AI workflow gave inconsistent results?",
        "How have you used Python, data, or AI tools in past projects?"
    ]

    strategy = (
        "The applicant should emphasize hands-on technical projects, ability to learn quickly, "
        "comfort with AI tools, documentation skills, and experience translating problems into working solutions."
    )

    return {
        "Role Summary": role_summary,
        "Key Skills Found": skills,
        "Resume Keywords": resume_keywords,
        "Interview Prep Questions": interview_questions,
        "Application Strategy": strategy
    }
```

## Sample Job Description

The sample job description below is based on an AI Agent Builder Intern role. The agent will analyze the description and generate structured career guidance.

```python
sample_job_description = """
AI Agent Builder Intern

Responsibilities include helping build intelligent AI agents, translating client and project requirements into actionable designs, experimenting with AI models and configurations, creating agent workflows, testing agents, documenting bugs, building integrations, writing guides and diagrams, using Git, and staying current with AI and machine learning systems.

Preferred skills include familiarity with AI tools, chatbots, conversational AI platforms, Python, JavaScript, APIs, webhooks, prompt writing, documents, spreadsheets, and knowledge bases.
"""
```

```python
results = analyze_job_description(sample_job_description)

for section, content in results.items():
    print(f"\n{section}")
    print("-" * len(section))

    if isinstance(content, list):
        for item in content:
            print(f"- {item}")
    else:
        print(content)
```
